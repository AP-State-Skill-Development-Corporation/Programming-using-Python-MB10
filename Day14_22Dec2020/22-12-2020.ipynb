- List Comprehension:
- by using list Comprehension we can create one new list from the other iterables.
- Syntax:
    - [output_expression for loop if(condition)]

```python
# 1. generate natural numbers from 1 to 20 using list?
#output : [1,2,3,4,5,6,............20]
l = []
for i in range(1,21):
    l.append(i)
print(l)
```

```python
[i for i in range(1,21)]
```

```python
# to print even numbers with in range of  1 to 100 using LC

k = [i for i in range(1,101) if i%2==0]
print(k)
```

```python
def odd(n):
    return n%2!=0
s = [i for i in range(1,101) if(odd(i))]
print(s)
```

```python
# to generate prime numbers with in the range of 1 to 100 using LC
prime_list = [x for x in range(0 ,100) for y in range(2,x) if x % x == 0 and x % 1 == 0 and x % y != 0]
print(prime_list)
```

```python
def prime(n):
    c=0
    for i in range(1,n+1):
        if n%i==0:
            c+=1
    if c==2:
        return n
prime1 = [i for i in range(1,101) if prime(i)]
print(prime1)
            
    
```

```python
n = 100

primes = [i for i in range(2, n + 1) if all(i%j != 0 for j in range(2, int(i ** 0.5) + 1))]

print(primes)
```

- Dictionary Comprehension:
    - syntax:
        - {key:value for loop if(condition)}

```python
# to generate squares of numbers between 1 to 10 in DC?
# output:{1:1,2:4,3:9,4:12,.....10:100}
d={}
for i in range(1,11):
    d[i]=i**2
print(d)    
```

```python
{i:i**2 for i in range(1,11)}
```

- lambda: 
     - by using "lambda" keyword we can develop the function.  
     - it is a anynomous function in a single line.
- syntax:
    - z = lambda arguments:expression
    - z(arguments)


```python
# to print addition of 2 numbers

a = lambda x,y,z:x+y+z
a(10,20,30)
```

```python
# print odd numbers using lambda
b = lambda x:x%2!=0
l = [i for i in range(1,100) if b(i)]
print(l)
```

- filter:
    - syntax:
        - filter(function_name,sequence)

```python
l = [10,-6,23,0,-3,-723,67,-8]
# print only negative values

def negative(n):
    return n<0
print(filter(negative,l))
print(tuple(filter(negative,l)))
print(list(filter(negative,l)))
print(set(filter(negative,l)))
```

- map:
    - syntax:
        - map(function_name,sequence)

```python
l = [10,-6,23,0,-3,-723,67,-8]
list(map(lambda n:n+10,l))
```

```python
a=list(map(int,"10 20 30".split()))
print(a)
```

```python
a = "10 20 30"
d = a.split()
print(d)
l=[]
for i in d:
    l.append(int(i))
print(l)
```

- Generator 
- to generate the sequence of values from the function using "yield" keyword

```python
def natural(n):
    while n<=5:
        return n
       
natural(4)
```

```python
def natural1(n):
    while n<=5:
        yield n
        n=n+1
       
list(natural1(1))
```
