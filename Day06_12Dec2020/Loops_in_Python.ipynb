Termination conditions

- Break - stops the execution of the current loop
- continue - The continue keyword is used to end the current iteration in a loop, and continues to the next iteration.


````python
for num in range(1, 50):
    if num == 5 or num == 10 or num == 15:
        continue
    elif num == 40:
        print("Supply is terminated")
        break
    else:
        print(num, end = ' ')
````
